```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("../data/nsl_kdd_dataset.csv")
```

```python
df.head(10)
```

```python
df['label']= df['label'].apply(lambda x:0 if x=='normal' else 1)
```

```python
df.head(10)
```

```python
df.groupby('label')['serror_rate'].mean()

```

```python
df.groupby('label')['count'].mean()
```

```python
df.groupby('label')['src_bytes'].mean()
```

```python
import matplotlib.pyplot as plt

normal = df[df['label'] == 0]['serror_rate']
attack = df[df['label'] == 1]['serror_rate']

plt.hist(normal, bins=30, alpha=0.6, label='Normal')
plt.hist(attack, bins=30, alpha=0.6, label='Attack')
plt.legend()
plt.title("Error Rate: Normal vs Attack")
plt.xlabel("serror_rate")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot([df[df['label']==0]['src_bytes'],
             df[df['label']==1]['src_bytes']],
            labels=['Normal','Attack'])
plt.title("Source Bytes: Normal vs Attack")
plt.ylabel("Bytes")
plt.show()
```

```python
plt.boxplot([df[df['label']==0]['count'],
             df[df['label']==1]['count']],
            labels=['Normal','Attack'])
plt.title("Connection Count: Normal vs Attack")
plt.show()
```

```python
corr = df.corr()['label'].sort_values(ascending=False)
corr.head(10)
```

 Security Insights

- Attack connections show significantly higher error rates
- Attacks involve abnormal data transfer behavior
- Attack traffic demonstrates aggressive connection repetition
- Certain features are highly correlated with intrusion activity
- Visual analysis confirms clear behavioral separation between attack and normal traffic
